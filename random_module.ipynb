2. 달(month)을   정수로   입력    받아   계절을   출력하는   프로그램을   작성하세요.

```python
month=int(input('달을 입력하세요.'))
if 3<=month<=5:
    print('%d월은 봄입니다'%month)
elif 6<=month<=8:
    print("%d월은 여름입니다"%month)
elif 9<=month<=11:
    print("%d월은 가을입니다"%month)
elif (month==12) or (1<=month<=2):
    print("%d월은 겨울입니다"%month)
else:
    print('잘못입력했습니다')
```
