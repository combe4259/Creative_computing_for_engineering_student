2.           . ,   10    .   ,  50000, 10000, 5000, 1000, 500, 100 , 50, 10 . 

```python
money=int(input(' :'))
cond=50000
change=0
while money>=10:
    if money>=cond:
        print("%d:%d"%(cond,money//cond))
        money-=cond*(money//cond)
    else:
        print("%d:%d"%(cond,0))
    if change==0:
        cond/=5
        change+=1
    else:
        cond/=2
        change-=1
    
```
