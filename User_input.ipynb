1. 사용자로부터 점수를 입력 받아 점수가 70점 이상이면 ‘합격입니다’, 70점 미만이면 ‘불합격입니다’를
출력하는 과정을 3번 반복하는 순서도를 작성하고, 이를 프로그램으로 코딩하세요.


```python
count=1;
while count<=3:
    point=int(input('점수를 입력하시오:'))
    if point>=70:
        print('합격입니다')
    else:
        print('불합격입니다')
    count=count+1
```

2. 요금을 입력 받아 각 화폐 단위의 필요 개수를 출력하는 프로그램을 작성하세요. 단, 화폐 단위가 10원이 될 때 까지만 반복한다. 요금은 정수로 입력되며, 화폐로는 50000원, 10000원, 5000원, 1000원, 500원, 100 원, 50원, 10원이 있다. 

```python
money=int(input('요금을 입력하세요:'))
cond=50000
change=0
while money>=10:
    if money>=cond:
        print("%d원:%d개"%(cond,money//cond))
        money-=cond*(money//cond)
    else:
        print("%d원:%d개"%(cond,0))
    if change==0:
        cond/=5
        change+=1
    else:
        cond/=2
        change-=1
    
```

피보나치 수열의 예:1 1 2 3 5 8 13

```python
count=1
a=1
b=1
n=int(input('항의 개수를 입력:'))
if n<=1:
        print("1",end=" ")
elif n<=2:
        print("1 1",end=" ")
else:
    print("1 1",end=" ")
while count<=n-2 and (n!=1 and n!=2):
    c=a+b
    print(c,end=" ")
    a=b
    b=c
    count=count+1
    
    
    
    
```
